### 1. Salary

```python
df['salary'].unique()
```

There seems to be some patterns

Try converting VND first

```python
def convert_salary_vnd(salary):
    if 'Trên' in salary:
        # Extract the value after "Trên" and convert it to integer
        value = int(re.search(r'\d+', salary.replace(',', '')).group())
        return value * 1000000  # Convert Tr to VND
    elif 'Lên đến' in salary:
        # Extract the value after "Lên đến" and convert it to integer
        value = int(re.search(r'\d+', salary.replace(',', '')).group())
        return value * 1000000  # Convert Tr to VND
    elif '-' and 'Tr' in salary:
        # Extract the range values, calculate their average, and convert to integer
        values = re.findall(r'\d+', salary.replace(',', ''))
        values = [int(value) for value in values]
        average_value = np.mean(values)
        return average_value * 1000000  # Convert Tr to VND
    else:
        # For values like "15,000,000 - 35,000,000 VND"
        values = re.findall(r'\d+', salary.replace(',', ''))
        values = [int(value) for value in values]
        if len(values) == 2:
            average_value = np.mean(values)
        else:
            average_value = values[0]
        return average_value
```

Next USD to VND, I use the exchange rate of 1 USD to 25,000 VND

```python
USD_TO_VND = 25000
def convert_salary_usd(salary):
    # Extract the value after "Trên" and convert it to integer
    if 'Trên' in salary:
        value = int(re.search(r'\d+', salary.replace(',', '')).group())
        return value * USD_TO_VND
    # Extract the value after "Lên đến" and convert it to integer
    elif 'Lên đến' in salary:
        value = int(re.search(r'\d+', salary.replace(',', '')).group())
        return value * USD_TO_VND
    # Extract the range values, calculate their average, and convert to integer
    elif '-' and 'Tr' in salary:
        values = re.findall(r'\d+', salary.replace(',', ''))
        values = [int(value) for value in values]
        average_value = np.mean(values)
        return average_value * USD_TO_VND
```

Combine them

```python
def convert_salary(salary):
    if 'Cạnh tranh' in salary:
        return None
    elif 'VND' in salary:
        return convert_salary_vnd(salary)
    elif 'USD' in salary:
        return convert_salary_usd(salary)
```

```python
df['salary'] = df['salary'].apply(convert_salary)
df['salary']
```

### 2. Experience

```python
df['experience'].unique()
```

They are ranged, i want to convert them to job level name (Entry-Level, Mid-Level, Senior-Level, Management-Level, Executive-Level, C-Suite)

```python
def job_level_classify(year):
    if year <= 2:
        return "Entry-Level"
    elif year <= 5:
        return "Mid-Level"
    elif year <= 10:
        return "Senior-Level"
    elif year <= 15:
        return "Management-Level"
    elif year <= 20:
        return "Executive-Level"
    else:
        return "C-Suite"
```

```python
def experience_to_single_job_level(experience):
    if experience is None:
        return None
    # No experince => Entry-level
    elif "Chưa có kinh nghiệm" in experience:
        return "Entry-Level"
    # If there is only 1 year in string, use that year
    elif "Trên" in experience:
        year = int(re.search(r'\d+', experience).group())
        return job_level_classify(year)
    elif "Lên đến" in experience:
        year = int(re.search(r'\d+', experience).group())
        return job_level_classify(year)
    # Else, use the max integer
    else:
        try:
            years = re.findall(r'\d+', experience)
            max_exp = int(years[1])
            return job_level_classify(max_exp)
        except:
            return None
```

```python
df['job_levels'] = df['experience'].apply(experience_to_single_job_level)
df[['experience', 'job_levels']]
```
